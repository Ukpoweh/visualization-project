## Step 3: Load the data

Use the next code cell to load your data file into `my_data`.  Use the filepath that you specified in the previous step.

````python
# Fill in the line below: Read the file into a variable my_data
my_data = pd.read_csv(my_filepath)

# Check that a dataset has been uploaded into my_data
step_3.check()
````

````python
#overview of the columns in the dataframe
my_data.info()
````

````python
weather = [' _conds', ' _fog', ' _hail', ' _rain', ' _snow', ' _thunder', ' _tornado']
for col in weather:
    print(my_data[col].value_counts())
````

Dropping the precipitation column since all its values are null

````python
my_data_2 = my_data.drop([' _precipm'], axis=1)
my_data_2
````

Converting the date column from object to datetime

````python
my_data_2['datetime_utc'] = pd.to_datetime(my_data_2['datetime_utc'])
my_data_2['datetime_utc'].dtype
````

````python
#overview of the new dataframe
my_data_2.info()
````

Creating a year column from the date column

````python
my_data_2['Year'] = my_data_2['datetime_utc'].dt.year
my_data_2['Year'].value_counts().sort_index(ascending=False)
````

The dataframe spans from the year 1996 to 2017.

## Step 4: Visualize the data

Use the next code cell to create a figure that tells a story behind your dataset.  You can use any chart type (_line chart, bar chart, heatmap, etc_) of your choosing!

````python
# Create a plot
sns.set()
plt.hist(x=' _dewptm', data=my_data_2, bins=50)# Your code here
plt.title("Frequency of the dewpoint column")

# Check that a figure appears below
step_4.check()
````

````python
sns.boxplot(my_data_2[' _pressurem'].dropna(axis=0))
````

The boxplot above reveals an outlier in the pressure column

````python
outlier = my_data_2[my_data_2[' _pressurem'] ==my_data_2[' _pressurem'].max()]
outlier
````

The row above is the row that contains the extreme value for outlier

````python
sns.boxplot(x=' _tempm', data=my_data_2)
plt.title('A boxplot showing the summary statistics of the temperature')
````

The temperature ranges from 0-90

````python
plt.hist(my_data_2[' _hum'].dropna(axis=0), bins=20)
plt.title('Frequency of the humidity')
````

````python
plt.figure(figsize=(16,20))
sns.countplot(y=my_data_2[' _conds'].sort_values(ascending=False))
````

From the countplot above, Haze is the most frequent weather condition

````python
my_data_2.columns
````

````python
sns.scatterplot(x=' _tempm', y=' _hum', data=my_data_2)
plt.title('Temperature vs humidity')
````

From the scatterplot above, there is no much correlation between the temperature and humidity

````python
temp_per_cond = my_data_2.groupby(' _conds')[' _tempm'].mean().sort_values(ascending=False).reset_index()
````

````python
plt.figure(figsize=(50,5))
sns.barplot(temp_per_cond[' _conds'], temp_per_cond[' _tempm']);
plt.xticks(rotation=90)
````

Widespread dust is the weather condition that has the highest average value for temperature

````python
temp_per_year = my_data_2.groupby('Year')[' _tempm'].mean().sort_values(ascending=False).reset_index()
temp_per_year
````

````python
plt.figure(figsize=(20,5))
sns.barplot(temp_per_year['Year'], temp_per_year[' _tempm'])
````

The highest average temperature value was recorded in 1999

````python
temp_per_dir = my_data_2.groupby(' _wdire')[' _tempm'].mean().sort_values(ascending=False).reset_index()
temp_per_dir
````

````python
my_data_2[' _wdire'].value_counts()
````

````python
sns.barplot(temp_per_dir[' _wdire'], temp_per_dir[' _tempm'])
plt.xticks(rotation=90)
````

Though the "variable"is the least frequent value in the column, it recorded the highest average temperature. This means it has significance and should not be dropped.
